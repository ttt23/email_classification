## Loading libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Loading and reading dataset

```python
df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
label_counts = df['Prediction'].value_counts()
label_percent = (label_counts/len(df))*100
# label_percent

plt.bar(label_percent.index, label_percent.values, color=['skyblue', 'salmon'])
plt.xlabel("Email Type")
plt.ylabel("Percentage (%)")
plt.title("Distribution of Ham and Spam Emails")
for i, v in enumerate(label_percent.values):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center')

plt.show()
```

## Splitting dataset

```python
iloc_indexer = df.iloc
X = iloc_indexer[:, 1:3001]
X
```

```python
y = iloc_indexer[:, -1].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

## Hard margin SVM

```python
svc = SVC(kernel='linear', C=1e7)
svc.fit(X_train, y_train)
print("Accuracy score of Hard marign SVM for training set:", accuracy_score(svc.predict(X_train), y_train))
print("Accuracy score of Hard marign SVM for testing set:", accuracy_score(svc.predict(X_test), y_test))
print('-----------------------------------------------------------------------------------------------------')
print("Precision score of Hard marign SVM for training set:", precision_score(svc.predict(X_train), y_train))
print("Precision score of Hard marign SVM for testing set:", precision_score(svc.predict(X_test), y_test))
print('-----------------------------------------------------------------------------------------------------')
print("Recall score of Hard marign SVM for training set:", recall_score(svc.predict(X_train), y_train))
print("Recall score of Hard marign SVM for testing set:", recall_score(svc.predict(X_test), y_test))
print('-----------------------------------------------------------------------------------------------------')
print("F1 score of Hard marign SVM for training set:", f1_score(svc.predict(X_train), y_train))
print("F1 score of Hard marign SVM for testing set:", f1_score(svc.predict(X_test), y_test))
```
