### 4. Tiền xử lý dữ liệu

Ở bước này, ta sẽ loại bỏ, tinh chỉnh các yếu tố (dấu câu, chữ viết hoa, các từ được đã được chia động từ...) ở phần *text* tại mỗi mẫu dữ liệu, để đảm bảo các *features* thu được sẽ dựa trên ngữ nghĩa, tránh hiện tượng *model* học dựa trên những *features* không mang nhiều ý nghĩa phân biệt về mặt ngữ nghĩa và tránh việc *model* sử dụng các *features* tương tự nhau về mặt ngữ nghĩa để phân loại. Ví dụ: nếu như ta không tinh chỉnh chữ viết hoa thành chữ viết thường, với trường hợp 2 chữ 't' và 'T', như vậy nếu như mà trong các ham email chỉ có chữ 't' nhỏ, còn trong các spam email chỉ có chữ 'T' lớn model hoàn toàn có thể sử dụng 2 *features* này để phân loại email. Và giả sử, ta có một ham email không có chữ 't' nhỏ nào, mà lại chứ nhiều chữ 'T' lớn thì khả năng cao nó sẽ được phân loại là spam, như vậy ta thấy nếu chúng ta dùng các *features* gần như không biệt nhau về mặt ngữ nghĩa hoặc là không có nhiều ý nghĩa về mặt ngữ nghĩa thì hiệu suất trên thực tế của *model* chắc chắn sẽ không tốt. 

Cụ thể, ở đây, ta sẽ tiến hành chuẩn hoá *text* về *lowercase*; xoá các kí tự đặc biệt, *stopwords*, *punctuation*; chuẩn hoá các từ đã được chia động từ về dạng chuẩn (e.g: playing -> play). 

```python
# Sử dụng để chuẩn hoá các từ đã chia động từ về dạng chuẩn; 
ps = PorterStemmer()

# Hàm tiền xử lý dữ liệu, làm các việc đã liệt kê ở trên; 
def transform_text(text):
    # Chuẩn hoá text về lowercase; 
    text = text.lower()
    
    # Lấy các word token để phục vụ cho việc lọc các ký tự đặc biệt
    # , stopwords, punctuation;
    text = nltk.word_tokenize(text)
    
    # Loại bỏ các ký tự đặc biệt; 
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
            
    # Loại bỏ stopwords, punctuation; 
    text = y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
        
    # Chuẩn hoá các từ đã chia động từ về dạng chuẩn;
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
    
    # Kết hợp những word token đã được xử lý vào một xâu string như ban đầu; 
    return " ".join(y)
```

Tiếp theo, ta sẽ tiến hành chuyển đổi (transform) *text*.

```python
df['trans_text'] = df['text'].apply(transform_text)
```

```python
df.head()
```

### 5. Áp dụng thuật toán

Đầu tiên, ta sẽ thực hiện trích chuẩn các *trans_text*, ở đây ta sẽ sử dụng TfidfVectorizer để tiến hành trích chuẩn. 

```python
X = TfidfVectorizer(max_features = 3000).fit_transform(df['trans_text']).toarray()
y = df['label'].values
```

```python
X, y
```

Tiếp theo, ta sẽ tách tập dữ liệu ban đầu ra thành các tập train và tập test. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train, y_train
```

```python
X_test, y_test
```

```python
# Use to draw precision_score, recall_score bar chart; 
model_names = []
precision_scores = []
recall_scores = []
```

#### Áp dụng thuật toán Soft margin SVM 

```python
svc = SVC(kernel='linear', C=100)
svc.fit(X_train, y_train)
y_pred_svc_train_set = svc.predict(X_train)
y_pred_svc_test_set = svc.predict(X_test)
model_names.append('SVC')
precision_scores.append(precision_score(y_test, y_pred_svc_test_set))
recall_scores.append(recall_score(y_test,y_pred_svc_test_set))
print("Accuracy score of Soft marign SVM for training set:", accuracy_score(y_train, y_pred_svc_train_set))
print("Accuracy score of Soft marign SVM for testing set:", accuracy_score(y_test,y_pred_svc_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Precision score of Soft marign SVM for training set:", precision_score(y_train, y_pred_svc_train_set))
print("Precision score of Soft marign SVM for testing set:", precision_score(y_test, y_pred_svc_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Recall score of Soft marign SVM for training set:", recall_score(y_train, y_pred_svc_train_set))
print("Recall score of Soft marign SVM for testing set:", recall_score(y_test,y_pred_svc_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("F1 score of Soft marign SVM for training set:", f1_score(y_train, y_pred_svc_train_set))
print("F1 score of Soft marign SVM for testing set:", f1_score(y_test, y_pred_svc_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Confusion matrix of Soft margin SVM for testing set:\n", confusion_matrix(y_test, y_pred_svc_test_set))
# Visualize confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,  y_pred_svc_test_set), display_labels=['Ham', 'Spam'] )
cm_display.plot()
plt.show()
```

#### Áp dụng thuật toán Naive Bayes

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
y_pred_mnb_train_set = mnb.predict(X_train)
y_pred_mnb_test_set = mnb.predict(X_test)
model_names.append('NB')
precision_scores.append(precision_score(y_test, y_pred_mnb_test_set))
recall_scores.append(recall_score(y_test, y_pred_mnb_test_set))
print("Accuracy score of NB for training set:", accuracy_score( y_train, y_pred_mnb_train_set))
print("Accuracy score of NB for testing set:", accuracy_score( y_test,y_pred_mnb_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Precision score of NB for training set:", precision_score(y_train, y_pred_mnb_train_set))
print("Precision score of NB for testing set:", precision_score(y_test, y_pred_mnb_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Recall score of NB for training set:", recall_score(y_train, y_pred_mnb_train_set))
print("Recall score of NB for testing set:", recall_score(y_test,y_pred_mnb_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("F1 score of NB for training set:", f1_score(y_train, y_pred_mnb_train_set))
print("F1 score of NB for testing set:", f1_score(y_test, y_pred_mnb_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Confusion matrix of Naive Bayes for testing set:\n", confusion_matrix(y_test, y_pred_mnb_test_set))
# Visualize confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,  y_pred_mnb_test_set), display_labels=['Ham', 'Spam'] )
cm_display.plot()
plt.show()
```

#### Áp dụng thuật toán Logistics Regression

```python
logi = LogisticRegression()
logi.fit(X_train, y_train)
y_pred_logi_train_set = logi.predict(X_train)
y_pred_logi_test_set = logi.predict(X_test)
model_names.append('LR')
precision_scores.append(precision_score(y_test, y_pred_logi_test_set))
recall_scores.append(recall_score(y_test, y_pred_logi_test_set))
print("Accuracy score of LR for training set:", accuracy_score( y_train, y_pred_logi_train_set))
print("Accuracy score of LR for testing set:", accuracy_score( y_test, y_pred_logi_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Precision score of LR for training set:", precision_score(y_train, y_pred_logi_train_set))
print("Precision score of LR for testing set:", precision_score(y_test, y_pred_logi_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Recall score of LR for training set:", recall_score(y_train, y_pred_logi_train_set))
print("Recall score of LR for testing set:", recall_score(y_test, y_pred_logi_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("F1 score of LR for training set:", f1_score(y_train, y_pred_logi_train_set))
print("F1 score of LR for testing set:", f1_score(y_test, y_pred_logi_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Confusion matrix of Logistic Regression for testing set:\n", confusion_matrix(y_test, y_pred_logi_test_set))
# Visualize confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,  y_pred_logi_test_set), display_labels=['Ham', 'Spam'] )
cm_display.plot()
plt.show()
```

#### Áp thuật toán k-NN

```python
# Normalize data;
scaler = StandardScaler()
# scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Implement model; 
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled, y_train)
y_pred_knn_train_set = knn.predict(X_train_scaled)
y_pred_knn_test_set = knn.predict(X_test_scaled)
model_names.append('kNN')
precision_scores.append(precision_score(y_test, y_pred_knn_test_set))
recall_scores.append(recall_score(y_test, y_pred_knn_test_set))
print("Accuracy score of kNN for training set:", accuracy_score( y_train,y_pred_knn_train_set))
print("Accuracy score of kNN for testing set:", accuracy_score( y_test, y_pred_knn_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Precision score of kNN for training set:", precision_score(y_train, y_pred_knn_train_set))
print("Precision score of kNN for testing set:", precision_score(y_test, y_pred_knn_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Recall score of kNN for training set:", recall_score(y_train, y_pred_knn_train_set))
print("Recall score of kNN for testing set:", recall_score(y_test, y_pred_knn_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("F1 score of kNN for training set:", f1_score(y_train, y_pred_knn_train_set))
print("F1 score of kNN for testing set:", f1_score(y_test, y_pred_knn_test_set))
print('-----------------------------------------------------------------------------------------------------')
print("Confusion matrix of kNN for testing set:\n", confusion_matrix(y_test, y_pred_knn_test_set))
# Visualize confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,  y_pred_knn_test_set), display_labels=['Ham', 'Spam'] )
cm_display.plot()
plt.show()
```

## III. Kết luận

Trước hết, ta sẽ vẽ biều đồ trực quan hóa 2 metrics: precision_score, accuracy_score.

```python
performance_df = pd.DataFrame({
    'Model': model_names,
    'Recall': recall_scores,
    'Precision': precision_scores
})

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Recall comparison
sns.barplot(data=performance_df, x='Recall', y='Model', hue='Model', palette='viridis', legend = False, ax=axes[0])
axes[0].set_title('Model Accuracy Comparison', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Accuracy', fontsize=12)
axes[0].set_ylabel('Model', fontsize=12)
axes[0].set_xlim([0.1, 1.0])  
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.3f', padding=3)
# Precision comparison
sns.barplot(data=performance_df, x='Precision', y='Model', hue='Model', palette='magma', legend = False, ax=axes[1])
axes[1].set_title('Model Precision Comparison', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Precision', fontsize=12)
axes[1].set_ylabel('Model', fontsize=12)
axes[1].set_xlim([0.1, 1.0])  
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.3f', padding=3)
plt.tight_layout()
plt.show()
```

Như vậy, ta thấy trong các mô hình thì Naive Bayes vẫn cho hiệu suất tốt nhất với precision: 0.988, và recall: 0.822.
