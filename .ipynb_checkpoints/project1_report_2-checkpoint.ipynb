```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import os
```

## Loading and reading dataset

```python
df = pd.read_csv('emails.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Prediction'].value_counts()
```

## Splitting dataset

```python
X = df.iloc[:, 1:3001]
X
```

```python
y = df.iloc[:,-1].values
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

## Hard margin SVM
Giá trị của C ở đây chỉ đơn giản là một giá trị vô cùng lớn, để đảm bảo việc thực hiện Hard margin classifier. Sẽ có giải thích ở phía dưới trong phần Soft margin SVM. 

```python
hard_svc = SVC(kernel = 'linear', C=1e6)
hard_svc.fit(x_train, y_train)
print("Accuracy score of Hard marign SVM for training set:", accuracy_score(hard_svc.predict(x_train), y_train))
print("Accuracy score of Hard marign SVM for testing set:", accuracy_score(hard_svc.predict(x_test), y_test))
```

## Soft margin SVM
Do Hard margin SVM có hiệu suất phân loại kém trong các trường hợp có nhiễu, hoặc không thể hoạt động trong trường hợp dữ liệu không *linearly separable* nên Soft margin SVM sẽ được sử dụng thường xuyên. Lưu ý: Soft margin SVM chỉ hoạt động hiệu quả với dữ liệu gần *linearly separable*.  
Nguyên tắc hoạt động của thuật toán Soft margin SVM là sẽ chấp nhận *missclasify* một vài điểm trong *training set* để thu được mặt phẳng phân loại tốt (có giá trị margin lớn), mang lại hiệu suất phân loại tốt hơn. Nhưng nếu *missclasify* quá nhiều điểm trong *training set* thì hiệu suất phân loại lại kém đi, điều này sẽ được quyết định bởi người thực hiện thông qua điều chỉnh tham số *regularization* C. Nếu C quá lớn hoặc tiến dần đến vô cùng thì thuật toán sẽ chỉ cho phép *missclasify* một số ít điểm, giá trị margin sẽ gần giống với trường hợp sử dụng Hard margin SVM; người lại nếu C quá nhỏ thì thuật toán cho phép *missclasify* nhiều điểm hơn, giá trị margin sẽ lớn hơn. Đây chính là lý do vì sao ở trên kia tôi đã đặt C = 1e6 - một giá trị rất lớn.  
Dưới đây, tôi sẽ tiến hành thực hiện Soft Margin SVM với 3 giá trị C khác nhau: C = 1e-6(rất nhỏ), C = 100(bình thường), C = 1e6(rất lớn)

```python
svc = SVC(kernel = 'linear', C=1e-6)
svc.fit(x_train, y_train)
print("Accuracy score of Soft marign SVM(C=1e-6) for training set:", accuracy_score(svc.predict(x_train), y_train))
print("Accuracy score of Soft marign SVM(C=1e-6) for testing set:", accuracy_score(svc.predict(x_test), y_test))
```

```python
svc = SVC(kernel = 'linear', C=100)
svc.fit(x_train, y_train)
print("Accuracy score of Soft marign SVM(C=100) for training set:", accuracy_score(svc.predict(x_train), y_train))
print("Accuracy score of Soft marign SVM(C=100) for testing set:", accuracy_score(svc.predict(x_test), y_test))
```

```python
svc = SVC(kernel = 'linear', C=1e6)
svc.fit(x_train, y_train)
print("Accuracy score of Soft marign SVM(C=1e6) for training set:", accuracy_score(svc.predict(x_train), y_train))
print("Accuracy score of Soft marign SVM(C=1e6) for testing set:", accuracy_score(svc.predict(x_test), y_test))
```
